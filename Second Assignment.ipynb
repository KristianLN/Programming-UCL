```python
%%javascript

///////////////////////////////////////////////////// Function library //////////////////////////////////////////////////////////
/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
///////////////////////////////////////////////////// Error Checking functions //////////////////////////////////////////////////

///////////////////////////////////////////////////// Check the ID provided /////////////////////////////////////////////////////

function checkid(id){
    // This function checks if the id only contains alphanumeric characters.
    // The function handles empty/undefined variables, spaces , no input ect.
  
    var check = new RegExp('^[a-zA-Z0-9]+$');
    var checkid=false,tempid;
    
    if (check.test(id)==true){
        
        // We are here if all characters are alphanumeric or undefined
        
        if (!!id==false){
            
            // we end here if the variable is undefined.
            // The two while loops below are identical and they give the user the option to reenter the id, and checks
            // that the new id only contains alphanumeric values and is not undefined.
            
            while(checkid==false){
                
                tempid = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' + 
                                'Please enter a ID containing only alphanumeric characters')
                
                if (check.test(tempid)==true){
                    
                    checkid=true;
                    
                    if (!!tempid==false){
                        
                        checkid=false;
                        
                    }
                }
            }
            // If the new id, provided by the user, only contains alphanumeric characters and is not undefined 
            
            return tempid
            
        } else {
            
            // if the variable only consists of alphanumeric characters and is not undefined
            
            return id
        }
        
    } else {
        
        // we end here if the initial id contains characters that aren't alphanumeric
        
        while(checkid==false){
            
            tempid = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                            'Please enter a ID containing only alphanumeric characters')
            
            if (check.test(tempid)==true){
                
                checkid=true;
                
                if (!!tempid==false){
                    
                    checkid=false;
                    
                }
            }
        }
        
        return tempid
    }
}

///////////////////////////////////////////////////// Check the capacity ////////////////////////////////////////////////////////

function checkcap(capacity){
    // This function checks the requirement underlying the input for capacity, 
    // namely that it is a numeric integer value.
    
    if (Number.isInteger(capacity)==true){
        
        return capacity
        
    } else {
        
        var tempcap;
        
        while(Number.isInteger(Number(tempcap))==false){
            
            tempcap = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                             'The capacity provided is not an integer, so please provide it again.')
            
        }
        
        return Number(tempcap)
        
    }
}

///////////////////////////////////////////////////// Check if parking fee apply /////////////////////////////////////////////////////

function checkparkingfee(parkingfee){
    // This function checks if the value for, whether parking fees applies or not, is boolean or string with boolean value.
    
    var values = ['True','true','false','False'];
    var checkpark = false,temppark;
    
    if (typeof(parkingfee)==='boolean' || (values.indexOf(parkingfee)>-1)){
        
        // We end here if the value is boolean or part of the 'string with boolean values' array (values)
        
        return parkingfee;
        
    } else {
        
        // we end here if the input is anything else than boolean or part of the 'string with boolean values' array (values)
        
        while(checkpark == false){
            
            temppark = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                              'Please enter whether a parking fee applies or not')
            
            if (typeof(temppark)==='boolean' || (values.indexOf(temppark)>-1)){
                
                checkpark = true;
            }
        }
        if (values.indexOf(temppark)<2){
            
            return true;
            
        } else {
            
            return false;
            
        }
    }
}

///////////////////////////////////////////////////// Check the Socket Type /////////////////////////////////////////////////////

function checksockettype(sockettype){
    //This function checks if the socket type is known, based on the list below.
    
    var types = ['Type2','type2','tesla_standard','Tesla_standard','Tesla_Standard','chademo','Chademo'];
    var types_reduced = ['Type2', 'Tesla_standard','Chademo'];
    var no = ['no','No'],yes = ['yes','Yes'];
    var tempsocket,addsocket,donito,nosockettype;
    var createdarray = [];
    var done = false,checksocket = false, wronginput = false;
    
    // The expected format of the input is an array, but below we check if the input is a string or an empty/undefined variable
    
    if (typeof(sockettype)=='string'){
        
        sockettype = [sockettype]
    
    } else if (!!sockettype==false){
        
        //Here we give the user the oppotunity to create a new array of socket types, if the input is empty.
        
        while(done==false){
            
            if (createdarray.length==0){
                
                nosockettype = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                                      'There is no value for the socket types, please enter one.')
                
            } else {
                
                nosockettype = prompt('Please enter another socket type.')
                
            }
            
            createdarray.push(nosockettype)
            
            if ((types.indexOf(nosockettype)>-1)==false){
                
                addsocket = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                                   'You entered an unknown socket type: '+ nosockettype +
                                   '. Is that a new type that should be contained in the list?')
                
                if (yes.indexOf(addsocket)>-1){
                
                // If yes, then we don't do anythign with the unknown type, except for adding it to the list of known types.
                
                    types.push(nosockettype)
                    types_reduced.push(nosockettype)
                }
            }
            
            donito = prompt('Do you want to add more socket types to the list of avaliable socket types for this charging station?')
            
            if (no.indexOf(donito)>-1){
            
                done = true;
            
            }
        }
        
        sockettype = createdarray;
    
    }
    
    for (var socket of sockettype){
        
        if ((types.indexOf(socket)>-1)){
            
            // we enter here if the socket type is part of the known list, and proceed without doing anything.
            
        } else {
            
            // The user has the option to add the new socket type to the list.
            
            while(wronginput==false){
            
                addsocket = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                                   'You entered an unknown socket type: '+ socket +
                                   '. Is that a new type that should be contained in the list?')

                    if (yes.indexOf(addsocket)>-1){

                        // If yes, then we don't do anythign with the unknown type, except for adding it to the list of known types.

                        types.push(socket);
                        types_reduced.push(socket);
                        wronginput = true;

                    } else if (no.indexOf(addsocket)>-1) {

                        // we enter here if the socket type is not known, and ask the user to provide new input.

                        while(checksocket==false){

                            tempsocket = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                                                'Please enter a type of socket that matches the known list, which are ' +
                                               types_reduced);

                            if (types.indexOf(tempsocket)>-1){

                                checksocket = true;

                                var ins = sockettype.indexOf(socket);

                                sockettype[ins] = tempsocket;

                            }
                        }
                        
                        wronginput = true;
                    
                    } else {
                        
                        prompt('I do not understand your answer. Please continue and give me your answer again.')
                        
                    }
            }
        }
    }
    
    return sockettype
}

///////////////////////////////////////////////////// Check the coordinates /////////////////////////////////////////////////////

function checkcoordinate(coordinate,min_coor,max_coor){
    // This function checks that the coordinates are of correct format. That implies checking that the values provided is non-empty, an integer and 
    // within the correct range, depending whether it is a x (longitude) or y (latitude) coordinate.
    // Longitude range: [-180,180] degrees
    // Latitude range: [-90,90] degrees
    
    var checkcoor = false,tempcoor,answer;
    var yes = ['yes', 'Yes'], no = ['no','No'];
    
    // The first thing to check is if the value is a numeric value, and lies in the correct span.
    
    if (!isNaN(coordinate) && (coordinate <= max_coor && coordinate >= min_coor)){
        
        // If that is the case, we check if it is an integer or a decimal value.
        
        if (Number.isInteger(coordinate)){
            
            // If it is an integer, we ask the user whether it is meant to be that way, or if it is an error.
            
            answer = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                            'The numeric value you gave me is an integer, do you want me to continue with it (yes), or is it an error (no)?');
            
            if (yes.indexOf(answer)>-1){
                
                return coordinate;
                
            } else if (no.indexOf(answer)>-1) {
                
                // If it is an error, we ask the user to provide a new coordinate
                
                while(checkcoor==false){
                    
                    tempcoor = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                                      'Please give me a new coordinate')
                    
                    // Again checking the format, as above
                    
                    if (!isNaN(tempcoor) && (tempcoor <= max_coor && tempcoor >= min_coor)){
                        
                        checkcoor = true;
                        
                        if (Number.isInteger(tempcoor)){
                            
                            console.log('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                                        'Ouch, the coordinate is not a float - try again')
                            
                            checkcoor = false;
                        }
                    }
                }
                
                return Number(tempcoor)
                
            } else {
                
                // we end up down here, if the answer to the question concerning the integer value doesn't make sense.
                // we ask the user to provide input we understand.
                
                //console.log('I can not understand your answer.')
                
                while(checkcoor==false){
                    
                    // If the user can answers us clearly, we will ask the user to provide us with a new coordinate input.
                    
                    tempcoor = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                                      'I can not understand your answer. Please give me a new coordinate. See the console for more information')
                    
                    // Checking the format again, as above
                    
                    if (!isNaN(tempcoor) && (tempcoor <= max_coor && tempcoor >= min_coor)){
                        
                        checkcoor = true;
                        
                        if (Number.isInteger(tempcoor)){
                            
                            console.log('Ouch, the coordinate is not a float - try again')
                            
                            checkcoor = false;
                            
                        }
                    }
                }
                
                return Number(tempcoor)
            }
                   
        }
        
    return coordinate
    
    } else {
        
        // we end up down here, if the input is not a numeric value e.g. a string, an empty variable ect.
        // we ask the user to provide input we understand.
        
        console.log('I can not understand the input for a coordinate you have provided me with. ' +
                   'Either you exceeded the limits for the coordinate or you gave me nonsense.')
        
        while(checkcoor==false){
            
            tempcoor = prompt('Error in establishing the ' + (stations.length+1) + 'th charging station. ' +
                              'Please give me a new coordinate, consult the console for more information.')
            
            //Checking the format of the input
            
            if (!isNaN(tempcoor) && (tempcoor <= max_coor && tempcoor >= min_coor)){
                
                checkcoor = true;
                
                if (Number.isInteger(tempcoor)){
                    
                    console.log('Ouch, the coordinate is not a float - try again')
                    
                    checkcoor = false;
                }
            }
        }
        
    return Number(tempcoor)
        
    }
}

///////////////////////////////////////////////////// Other Functions //////////////////////////////////////////////////////////////
////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

///////////////////////////////////////////////////// Charging Station Object /////////////////////////////////////////////////////

function ChargingStation(id, capacity, parkingfee,socket,coordinate_x,coordinate_y){
    
    this.id = checkid(id);
    this.capacity = checkcap(capacity);
    this.parkingfee = checkparkingfee(parkingfee);
    this.socket = checksockettype(socket);
    this.coordinate_x = checkcoordinate(coordinate_x,-180,180);
    this.coordinate_y = checkcoordinate(coordinate_y,-90,90);
    this.report = function(){
        console.log('| Charging Station ID: ' + this.id,
                    '| Capacity: ' + this.capacity,
                    '| Is free parking avaliable?: ' + this.parkingfee,
                    '| Types of sockets at this station: ' + this.socket,
                    '| Location of the station: ' + '('+[this.coordinate_x,this.coordinate_y]+')',
                    '|')
    }
    stations.push(this)
}

///////////////////////////////////////////////////// Location of the stations with free parking /////////////////////////////////////////////////////

function locationsoffreepark(array){
    // This function extracts the ID and location of the parking stations with free parking.
    
    var ele,item;
    var i = 1;
    var stationswithfreepark = [['ID',['X Coordinate','Y Coordinate']]];
    
    for (ele of array){
        
        if (ele.parkingfee == true){
            
            stationswithfreepark.push([ele.id,[ele.coordinate_x,ele.coordinate_y]])
        
        }
    }
    
    if (stationswithfreepark.length >1){
        
        for (item of stationswithfreepark){
            
            console.log('|    ' + item[0] + '    |' + '  [' + item[1] + ']  |' )
        
        }    
    }
}

///////////////////////////////////////////////////// Determining the Geographic Span /////////////////////////////////////////////////////

function geographicextent(array){
    // This function calculates the geographic span of the stations by location the maximum and minimum
    // values of the coordinates
    
    var item;
    var x_min = 180, y_min = 90;
    var x_max = -180, y_max = -90;
    
    if (array.length>0){
        
        for (item of array){
            
            if (item.coordinate_x > x_max){
                
                x_max = item.coordinate_x
            }
            
            if (item.coordinate_x < x_min){
                
                x_min = item.coordinate_x
            }
            
            if (item.coordinate_y > y_max){
                
                y_max = item.coordinate_y
            }
            
            if (item.coordinate_y < y_min){
                
                y_min = item.coordinate_y
            }
        
        }
        
        console.log('The maximum x coordinate is ' + x_max);
        console.log('The minimum x coordinate is ' + x_min);
        console.log('The maximum y coordinate is ' + y_max);
        console.log('The minimum y coordinate is ' + y_min);
    
    } else {
        
        console.log('Error - no area defined');
    
    } 
}

///////////////////////////////////////////////////// Determining the Total Capacity of the City /////////////////////////////////////////////////////

function exploringcapacity(array){
    // This function sums the total capacity in the city
    
    var ele;
    var output = 0;
    
    for (ele of array){
        
        output += Number(ele.capacity);
    
    }
    
    console.log('The total capacity in the city is: ' + output);
}

///////////////////////////////////////////////////// Capitalize First Letter /////////////////////////////////////////////////////
///////////////////////////////////////////////////// Borrowed from Steve Harrison [Reference at the bottom] //////////////////////

function capitalizeFirstLetter(string) {
    return string.charAt(0).toUpperCase() + string.slice(1);
}

///////////////////////////////////////////////////// Locate the Subset Based on Socket Type /////////////////////////////////////////////////////

function locatespecialsocket(array,sockettype){
    // This function takes two inputs, an array and a socket type. The array is searched for the specific 
    // socket type, and a subset is returned.
    
    var item,socket,ele;
    var subset = [];
    
    for (item of array){
        
        socket = item.socket;
        
        if (socket.indexOf(sockettype) > -1 || socket.indexOf(capitalizeFirstLetter(sockettype)) > -1){
            
            subset.push(item);
        
        }
    }
    
    if (subset.length >0){
        
        console.log('The subset of stations that have the socket type ' + sockettype + ' is: ')
        console.log('')
        
        for (ele of subset){
            
            console.log(ele);
            
        }
    
    } else {
        
        console.log('There are no charging stations with that socket type.')
    
    }
    console.log('')
    return subset;   
}

/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
///////////////////////////////////////////////////// Task 1 ///////////////////////////////////////////////////////////////////////
////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

var stations = [];
var station,subset_stations;

console.log('| OUTPUT |')

///////////////////////////////////////////////////// Create all the objects you need, I'll keep track of them ///////////////////

// Not giving any input allows the user the specify everything themselves - try it out!

// Uncomment to create the objects!

//var chargingstation_1 = new ChargingStation()
//var chargingstation_2 = new ChargingStation('Ag52680',5,true,'type2',51.5403,1.0403)
//var chargingstation_3 = new ChargingStation('A87223280',2,true,'chademo',-170.00,87.0403)
//var chargingstation_4 = new ChargingStation('AdfbbA2080',10,true,'type2',51.5403,1.0403)

///////////////////////////////////////////////////// Alternative way of providing input ///////////////////////////////////////////
///////////////////////////////////////////////////// Not necessarly more elegant //////////////////////////////////////////////////

/*
var input = [['Xsdfsd',90,true,['type2','chademo'],17.9870,80.540],
            ['Xsdfsd',90,true,['type2','chademo'],17.9870,80.540],
            ['Xsdfsd',90,true,['type2','chademo'],17.9870,80.540],
            ['Xsdfsd',90,true,['type2','chademo'],17.9870,80.540],
            ['Xsdfsd',90,true,['type2','chademo'],17.9870,80.540],
            ['Xsdfsd',90,true,['type2','chademo'],17.9870,80.540],
            ['Xsdfsd',90,true,['type2','chademo'],17.9870,80.540]]

function createstations(array){
    
    var chargingstation;
    var int;
    
    for (int of array){
        
        chargingstation = new ChargingStation(int[0],int[1],int[2],int[3],int[4],int[5]);
        
    }
    
}

createstations(input);
*/

///////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

console.log('')
console.log('You have at this point created '+ stations.length + ' objects succesfully, with no errors in the attribute values.')
console.log('')
console.log('The charging stations created are:')
console.log('')

for (station of stations){
    
    console.log(station)
}

/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
///////////////////////////////////////////////////// Task 2 ///////////////////////////////////////////////////////////////////////
////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

console.log('')
console.log('| LOCATION OF THE FREE PARKING |')
console.log('')

locationsoffreepark(stations)

/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
///////////////////////////////////////////////////// Task 3 ///////////////////////////////////////////////////////////////////////
////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

console.log('')
console.log('| GEOGRAPHIC EXTENT |')
console.log('')

geographicextent(stations)

/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
///////////////////////////////////////////////////// Task 4 ///////////////////////////////////////////////////////////////////////
////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

console.log('')
console.log('| EXPLORING CAPACITY |')
console.log('')

exploringcapacity(stations)

/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
///////////////////////////////////////////////////// Task 5 ///////////////////////////////////////////////////////////////////////
////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

console.log('')
console.log('| AN APPLIED CASE |')
console.log('')

subset_stations = locatespecialsocket(stations,'chademo')

geographicextent(subset_stations)

console.log('')

///////////////////////////////////////////////////// RERFERNCE ////////////////////////////////////////////////////////////////
///////////////////////////////////////////////////// https://stackoverflow.com/questions/1026069/how-do-i-make-the-first-letter-of-a-string-uppercase-in-javascript /////////////////////////////////////////////////////
```
