```python
%%javascript

// CELL 3
// find the variation of the data and return it

// First, the calculate the mean of the data 

function mean(data){
    
    var sumofdata = 0;
    
    for (var ele of data){
        
        sumofdata += parseInt(ele)
    }
    
    return sumofdata/data.length
}

function findVariance(data){
    
    var mean_data = mean(data);
    var ssd = 0;
    
    // Calculate the Sum of Squared Differences
    
    for (var ele of data){
        
        ssd += (parseInt(ele)-mean_data)**2;
    }
    
    // Return the population* variance
    // *Use (n-1) for the sample variance
    
    return ssd/data.length ;
}

var myVar = findVariance([1,2,3,4,5]);
IPython.notebook.kernel.execute('variance=' + myVar + ';');
```

The code handles multi-header-rows, missing values in rows and allows the user to convert data to a user-specified type. The only thing that can't be done so far, is that the user manually can specify the values there appear to be missing.

```python
### CSV and txt files are processed without problems and you provide other formats, you will be asked kindly to try again.
### However, all of this is quite redundant to do, as one properly just would use the amazing Pandas library in production.

import re

##############################################################################################################################
########################################### The get-colmun-of-interest function ##############################################

### Intelligent way of locating the column of variable of choice, this is useful below.

def getcolindex(var,header):
       
    ### The intelligent function assumes that only one variant of the variable of choice is present in the datafile,
    ### which of cause could be extended, but this seems already as above and beyond.
    
    i = 0
    
    for x in header:
        
        if var in x:
            
            varcol = i
        
        i += 1
    
    return varcol

##############################################################################################################################
###################################################### The dataset format function ###########################################

### Function to check if format is of 'correct' type

def EnsureCorrectFormat(file):
    
    knowntypes = ['csv','txt']
    useablefile = False
    
    if ('csv' in file):
        
        print('It is a CSV file we are working with!')
        print()
        
        pattern = re.compile("^\s*|\s*,|\s*$")
        useablefile = True
    
    elif ('txt' in file):
        
        print('It is a text file we are working with!')
        print()
        
        pattern = re.compile("^\s*|\s*\t|\s*$")
        useablefile = True
        
    else:
        
        ### if the format of the filetype isn't known, the user is asked to try again
        
        while(useablefile == False):
            
            filenam = input('I do not know how to process the filetype you gave me, will you please provide me with CSV or txt file. New file: ')
            print()
            
            if any(types in filenam for types in knowntypes):
                
                if ('csv' in filenam):
        
                    print('Now it is a CSV file we are working with, how delighted!')
                    print()
                    pattern = re.compile("^\s*|\s*,|\s*$")
                
                elif ('txt' in filenam):
        
                    print('Now it is a text file we are working with, how delighted!')
                    print()
                
                    pattern = re.compile("^\s*|\s*\t|\s*$")
            
                useablefile = True
    
    ### At this point, the user has provided input of known type, and we can continue.
    
    try:
        return filenam,pattern
    except:
        return file,pattern

##############################################################################################################################
################################################ The missing value function ##################################################
    
def findmissingvalues(data):

    count = {}
    initial = 0
    
    for ele in data:

        if str(len(ele)) not in count.keys():
            
            initial = 0
            count[str(len(ele))] = initial + 1
    
        initial = count[str(len(ele))] 
        
        count[str(len(ele))] = initial + 1
    
    maxkey = [x for x,y in count.items() if y == max(count.values())]
            
    return maxkey,len(count)
    
##############################################################################################################################
###################################################### The data wiz function #################################################
    
# The purpose of this function is to remove all header-look-a-like rows, which could occur if dataset for different years had 
# been merged together by rows. At the end of the function below, all numeric values are converted to a type of choice.

def dataWiz(data,typeof):
    
    corr = len(data)
    keeptrack,datatypes = {},{}
    yes = ['yes','Yes']
    no = ['no', 'No']
    
    indexinkeeptrack = 0
    
    # Check all the type of the data, output: 'datatypes' which is returned at the end - take a look.
    
    for liist in data:
    
        typeoftype,strtype = 0,0
        individualdatatype = {}
        
        for ele in liist:
            
            try:
                
                typeof(ele)
                typeoftype +=1
            
            except:
            
                strtype += 1
                
        individualdatatype['typeoftype'] = typeoftype
        individualdatatype['strtype'] = strtype
        
        datatypes[str(indexinkeeptrack)] = individualdatatype
        
        indexinkeeptrack += 1
        
    correction = len(data)
    
    for pos in datatypes:
        
        # If the row appears to be a header row, the user can store it and remove it from the data.
        
        if sum(datatypes[pos].values()) == datatypes[pos]['strtype']:
            
            if 'header' not in locals():
                
                print('')
                yesplease = input('The ' + str(int(pos)+1) + 'th row appears to be a header row. ' +
                                  'Namely ' + str(data[int(pos)]) +
                                  '. Do you want to store it as a header row, and thereby remove it from the data? ')
            
                if yesplease in yes:
                    
                    header = data[int(pos)]
                    keep = pos
                    del data[int(pos)]
                    
                if yesplease in no:
                    
                    overwrite = input('Do you want to remove the row anyway?')
                    
                    if overwrite in yes:
                        
                        del data[int(pos)]
                
                # Now the user gets the opportunity to remove all rows which appears to be header rows automatically,
                # instead of going through them one by one.
                print('')
                removeall = input('Now that one header row has been detected, would you like to remove all rows that appears to be headers?' + 
                                  ' "Yes" to remove all similar rows or "No" to go through them manually. ')
                
                if (removeall in yes):
                   
                    for posi in datatypes:
                        
                        # It is necessary to correct for the fact that a header row is removed.
                        
                        correction_2 = len(data)
                        adjust = correction - correction_2
                        
                        if posi != keep and sum(datatypes[posi].values()) == datatypes[posi]['strtype']:
                            
                            del data[int(posi)-adjust]
                    break
                    
            # We come done here if the user chose not to to remove all header-look-a-like rows automatically.
                    
            else:
                
                # It is necessary to correct for the fact that a header row is removed.
                
                correction_3 = len(data)
                adjust_2 = correction-correction_3
                
                print('')
                overwrite = input('There has already been detected one header row, do you want to overwrite it?'
                                 + 'The new header candidate is' + str(data[int(pos)-adjust_2]))
                    
                if (overwrite in yes):
                    
                    header = data[(int(pos)-adjust_2)]
                    del data[(int(pos)-adjust_2)]
                        
                elif (overwrite in no):
                    
                    print('')
                    dele = input('Do you want to remove the row anyway, that appeared to be a header' + 
                                 ' which was not chosen as header? ')
                        
                    if dele in yes:
                            
                        del data[int(pos)-adjust_2]
    
    # The user gets the opportunity to convert all numeric values to a type of choice (float or int)
    
    print('')
    convert = input('The header column is chosen to be: ' + str(header) + 
                    ' Do you want to convert all possible values to the type of choice? '+
                   'Note: the choice of type is determine by the input in "convert" variable (float default)')
    
    if convert in yes:
        
        l = 0
        
        for liist in data:
    
            p = 0
            
            for ele in liist:
            
                try:
                
                    data[l][p] = typeof(ele)
                
                except:
                    
                    None
                
                p += 1
                
            l += 1
            
    return header,datatypes,data

##############################################################################################################################
###################################################### The mother function ###################################################

def readInMyFile_extended(filename,variable = 'pop', convert = float):
    
    accessmode = 'r'
    middleman = EnsureCorrectFormat(filename)
    fileformat = middleman[0]
    pat = middleman[1]
    
    with open(fileformat, accessmode) as Myfile:
        
        data = [[x for x in pat.split(line) if x] for line in Myfile]
        data = [cleaneddata for cleaneddata in data if cleaneddata]    
    
    # Raising a concern if the dataset contains rows of different lengths
    
    middleman_3 = findmissingvalues(data)
    
    if middleman_3[1] > 1:
        
        print('')
        aware = input('The dataset appears to contain rows of different lengths. All rows with different length than the majority '
                     + 'is removed. It will be possible to correct the error by the user in a future extension. For now, press any key to continue. ')    
        
        data = [x for x in data if len(x)==int(middleman_3[0][0])]

    # Correcting for multiple header-rows and converting data 
        
    middleman_2 = dataWiz(data,convert)
    
    head = middleman_2[0]
    data = middleman_2[2]
    
    ### Intelligent way of getting the column of choice, with the default being population.

    varcolumn = getcolindex(variable,head)

    ### Extract all the present values of the variable of choice.
    ### We check if the type of variable can be used for variance calculation, float is used because the calculations
    ### doesn't care about float or integer.
    
    try:

        float(data[0][varcolumn])
    
    except:
        
        wrongoptionalvar = False
        
        while(wrongoptionalvar==False):
            
            print('')
            optionalvar = input('The variable you have specified is not a type that can be calculated with. Please give me another variable.')
            
            varcolumn = getcolindex(optionalvar,head)
            
            if type(data[0][varcolumn])!=str:
        
                wrongoptionalvar = True
                 
    varofchoice = [float(linefromcsv[varcolumn]) for linefromcsv in data]
    
    print('')
    print('Choose 0 for the raw cleaned data, pick 1 for the header of the data, 2 for the pop and 3 for variance of the pop data')
    print()
    print('The error message appears because the RegExp is used on the non-cleaned data,',
          'which is done to keep the kode as tight as possible, and because nested list comprehensions',
          'is kinda bad ass.')
    
    ### Calculate variance of the variable of choice.
    
    variance = sum( ( [ ( numb - ( sum(varofchoice) / len(varofchoice) ) ) **2 for numb in varofchoice ] ) ) / len(varofchoice)
    
    return data,head,varofchoice,variance   
```
